# Now we will apply Count or Text vectorization but without stemming

__CountVectorizer is a feature extraction technique used in natural language processing (NLP) to transform a collection of text documents into a matrix of word counts__

__The CountVectorizer works by tokenizing each document, i.e., splitting it into individual words (or tokens), and then counting the frequency of each word in each document. The output is a matrix where rows represent the documents in the corpus, columns represent the unique words in the corpus, and each cell contains the count of the word in the corresponding document. This is also known as a "bag-of-words" representation.__

__CountVectorizer can be useful for a wide range of NLP tasks, including text classification, sentiment analysis, topic modeling, and more. It is a simple and effective way to represent text data numerically, which can then be used as input to machine learning algorithms.__

__*Here's an example of how to use CountVectorizer in Python using scikit-learn:*__

# Now we will Applying stemming:

## what is stemming ?

__Ans:__

__Stemming is a text preprocessing technique used to reduce words to their base or root form by removing any affixes (suffixes or prefixes) that are attached to the end of the word. For example, stemming the word "jumping" would result in "jump", and stemming the word "happily" would result in "happy".__

Let's say we have three words: jumping, jumps, and jumped. These words all have the same root word, which is "jump". Stemming is the process of reducing these words to their common root form, or stem.

After stemming, the three words would become:
<pre>
jumping → jump
jumps → jump
jumped → jump
</pre>
As we can see, stemming reduces different forms of the same word to a common root form. This can be useful in natural language processing tasks where we want to treat different forms of the same word as equivalent, such as in text classification or sentiment analysis. By applying stemming, we can reduce the number of unique words in a corpus and make it easier to analyze the text data.

__In the Scikit-learn (sklearn) library, stemming can be performed using the "Porter stemmer" algorithm. The Porter stemmer is a widely used algorithm for stemming in English. The algorithm works by applying a series of rules to reduce a given word to its stem.__

__The Scikit-learn library provides a class called "CountVectorizer" that can be used for text preprocessing tasks, including stemming. The CountVectorizer class has a parameter called "tokenizer" that can be set to a function that performs stemming.__

__Here is an example of how to perform stemming using the CountVectorizer class in Scikit-learn:__

*python code*

<pre>
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
def tokenize(text):
    words = nltk.word_tokenize(text)
    stemmed_words = [stemmer.stem(word) for word in words]
    return stemmed_words
vectorizer = CountVectorizer(tokenizer=tokenize)

</pre>

__In this example, the "PorterStemmer" from the <span style="color:green">*Natural Language Toolkit (NLTK) library*</span> is used to perform stemming, and the "tokenize" function is passed to the CountVectorizer class as the tokenizer parameter to perform stemming on the input text.__

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# example
ps.stem('loved')
```

```python
def steming(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

# Now we will find the cosine similarity between the vectors

## what is a cosine similarity ?

## Ans:



The cosine similarity is a measure of similarity between two non-zero vectors of an inner product space. It is commonly used in machine learning to measure the similarity between documents or feature vectors.

In scikit-learn (also known as sklearn), the cosine similarity can be calculated using the cosine_similarity function from the metrics module. Here's an example of how to use it:

## python code

<pre>
<span style="color:Green">from sklearn.metrics.pairwise import cosine_similarity</span>

# Example vectors
vector1 = [1, 2, 3]
vector2 = [4, 5, 6]

# Calculate the cosine similarity between the two vectors
cosine_sim = cosine_similarity([vector1, vector2])

print(cosine_sim)
</pre>

## output

<pre>
The output of the above code will be a 2x2 matrix containing the cosine similarity between the two input vectors:


[[1.         0.97463185]
 [0.97463185 1.        ]]
 
</pre>

The diagonal values are always 1, since the cosine similarity between a vector and itself is 1. The off-diagonal values are the cosine similarity between the two input vectors. In this example, the cosine similarity between vector1 and vector2 is 0.975.
